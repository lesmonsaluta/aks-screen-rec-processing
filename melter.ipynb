```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
import matplotlib.pyplot as plt
import warningsx
warnings.filterwarnings('ignore')


```

```python
wifi413 = pd.read_csv('4_13_wifi.csv')
data413 = pd.read_csv('4_13_data.csv')
wifi414 = pd.read_csv('4_14_wifi.csv')
data414 = pd.read_csv('4_14_data.csv')
```

```python
wifi413
```

```python
melted_wifi413 = pd.melt(wifi413)
melted_wifi413.to_csv('melted_wifi413.csv', index=False)
melted_wifi413['Table'] = 'wifi413'
melted_wifi413
```

```python
melted_wifi414 = pd.melt(wifi414)
melted_wifi414.to_csv('melted_wifi414.csv', index=False)
melted_wifi414['Table'] = 'wifi414'
melted_wifi414
```

```python
melted_data413 = pd.melt(data413)
melted_data413
melted_data413['Table'] = 'data413'
melted_data413.to_csv('melted_data413.csv', index=False)
```

```python
melted_data414 = pd.melt(data414)
max_index = melted_data414['value'].idxmax()
melted_data414.loc[max_index, 'value'] = 1

melted_data414['Table'] = 'data414'

melted_data414.to_csv('melted_data414.csv', index=False)
melted_data414
```

```python
max_row = melted_data414.loc[melted_data414['value'].idxmax()]
print("Row with maximum value:")
print(max_row)
```

```python
melted_data413
```

```python
combined_data = pd.concat([melted_data413, melted_data414, melted_wifi413, melted_wifi414])
```

```python
colors = sns.color_palette('bright', 12).as_hex()
colors[0]
```

```python

  
# use to set style of background of plot
sns.set(style = 'whitegrid')

plt.figure(dpi=300)
sns.stripplot(x='variable', y='value', hue='Table', data=combined_data, jitter=0.25, size=2, palette=colors[8:12])
# Set the y-axis label
plt.ylabel('Time')

# Set the x-axis label
plt.xlabel('Timestamps')

# Display the plot
plt.show()
```
